```python
def fit_model(x_train,y_train,x_test,y_test):
    models = {
        'Random Forest': rnd,
        'Gradient Boosting': gbc,
        'Extra Trees': etc,
        'Hist Gradient Boosting': hgbc
    }
    
    for name, model in models.items():
        model.fit(x_train, y_train)
        accuracy = model.score(x_test, y_test)
        print(f"{name}--- Accuracy: {accuracy:.2f}")
    
```

```python
from collections import defaultdict

def get_threshold_values(clf,x,y,class_label,feature_names):
    class_samples =x[y==class_label]
    node_indicator = clf.decision_path(class_samples)
    leaf_ids = clf.apply(class_samples)

    feature = clf.tree_.feature
    threshold = clf.tree_.threshold
    feature_thresholds = defaultdict(list)
    for sample_idx in range(class_samples.shape[0]):
        node_index = node_indicator.indices[
            node_indicator.indptr[sample_idx]: node_indicator.indptr[sample_idx+1]
        ]
        for node_id in node_index:
            if feature[node_id]!=-2: #skip leaf nodes
                fname= feature_names[feature[node_id]]
                thresh = threshold[node_id]
                feature_thresholds[fname].append(thresh)

    # aggregate: return the min-max values for each class
    summary = {}
    for feat,values in feature_thresholds.items():
        summary[feat] = np.median(values)
    
    return summary
```

```python
def get_rf_threshold_values(rf_clf, x, y, class_label, feature_names):
    class_samples = x[y == class_label]
    feature_thresholds = defaultdict(list)

    for tree in rf_clf.estimators_:
        node_indicator = tree.decision_path(class_samples)
        feature = tree.tree_.feature
        threshold = tree.tree_.threshold

        for sample_idx in range(class_samples.shape[0]):
            node_index = node_indicator.indices[
                node_indicator.indptr[sample_idx]: node_indicator.indptr[sample_idx + 1]
            ]
            for node_id in node_index:
                if feature[node_id] != -2:  # skip leaves
                    fname = feature_names[feature[node_id]]
                    thresh = threshold[node_id]
                    feature_thresholds[fname].append(thresh)

    # Aggregate: min, max, median
    #index_to_label.get(target_class)
    summary={}
    for feat, values in feature_thresholds.items():
        summary[feat] = {
            'min': float(np.min(values)),
            'max': float(np.max(values)),
            'optimum': float(np.median(values))
        }
    # convert to DataFrame for better readability
    summary_df = pd.DataFrame.from_dict(summary, orient='index')
    summary_df.columns = ['Min', 'Max', 'Optimum']
    summary_df.index.name = 'Feature'

    crop_name = index_to_label.get(target_class)
    summary_df.columns = pd.MultiIndex.from_product([[crop_name], summary_df.columns])
    return summary_df

    # return summary
```

```python

def get_rf_threshold_values(rf_clf, x, y, index_to_label, feature_names):
    # Initialize an empty dictionary to hold results
    crop_feature_summary = {}

    # Loop through all the crops (target classes)
    for target_class in np.unique(y):
        class_samples = x[y == target_class]
        feature_thresholds = defaultdict(list)

        # Loop through each tree in the random forest
        for tree in rf_clf.estimators_:
            node_indicator = tree.decision_path(class_samples)
            feature = tree.tree_.feature
            threshold = tree.tree_.threshold

            # Loop through each sample in the class
            for sample_idx in range(class_samples.shape[0]):
                node_index = node_indicator.indices[
                    node_indicator.indptr[sample_idx]: node_indicator.indptr[sample_idx + 1]
                ]
                for node_id in node_index:
                    if feature[node_id] != -2:  # skip leaf nodes
                        fname = feature_names[feature[node_id]]
                        thresh = threshold[node_id]
                        feature_thresholds[fname].append(thresh)

        # Aggregate min, max, and median for each feature
        crop_summary = {}
        for feat, values in feature_thresholds.items():
            crop_summary[feat] = {
                'min': float(np.min(values)),
                'max': float(np.max(values)),
                'optimum': float(np.median(values))
            }

        # Add the summary for this crop to the dictionary
        crop_name = index_to_label.get(target_class)
        crop_feature_summary[crop_name] = crop_summary

    return crop_feature_summary
```
