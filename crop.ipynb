```python
crop =crop.sample(frac=1, random_state=42).reset_index(drop=True)
crop.head()
```

```python
from sklearn.preprocessing import LabelEncoder
unique_crop = crop['Crop'].unique()
le= LabelEncoder()
unique_crop= le.fit_transform(unique_crop)
crop_map=dict(zip(le.classes_, unique_crop))
crop_map
```

```python
see=le.get_params(deep=True)
see
```

```python
import pickle
with open('crop_map.pkl', 'wb') as f:
    pickle.dump(crop_map, f)
```

```python
crop['crop_no'] = crop['Crop'].map(crop_map)
crop.head()
```

```python
crop = crop.drop(columns=['Crop'])
crop.head()
```

```python
train_data= crop.iloc[:1980] # 80% of the data
test_data= crop.iloc[1980:] # 10% of the data
train_data.shape, test_data.shape
```

```python
from sklearn.model_selection import train_test_split
# Splitting the data into features and target variable
X = train_data.drop(columns=['crop_no'])
y = train_data['crop_no']
x_train,x_test,y_train,y_test = train_test_split(X,y,train_size=0.8,random_state=42)

x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, ExtraTreesClassifier, HistGradientBoostingClassifier
rnd= RandomForestClassifier()
gbc= GradientBoostingClassifier()
etc= ExtraTreesClassifier()
hgbc= HistGradientBoostingClassifier()
```

```python
def fit_model(x_train,y_train,x_test,y_test):
    models = {
        'Random Forest': rnd,
        'Gradient Boosting': gbc,
        'Extra Trees': etc,
        'Hist Gradient Boosting': hgbc
    }
    
    for name, model in models.items():
        model.fit(x_train, y_train)
        accuracy = model.score(x_test, y_test)
        print(f"{name}--- Accuracy: {accuracy:.2f}")
    
```

```python
compare_model = fit_model(x_train,y_train,x_test,y_test)
compare_model
```

```python
from sklearn.metrics import  classification_report,confusion_matrix,ConfusionMatrixDisplay
```

```python
rnd.fit(x_train,y_train)
y_pred = rnd.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=rnd.classes_)
disp.plot(cmap='summer') 
plt.title('Random Forest Classifier Confusion Matrix')
plt.show()
```

- the model (random forest classifier) has really learnt well on the data; not just mere memorizing but learning the data:
   - Having only one wrong prediction on 396 cases is impressive
